```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mousedf = pd.read_csv(mouse_drug_data_to_load)
clinicaldf = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
bothdf = mousedf.merge(clinicaldf, on='Mouse ID', how='outer')

# Display the data table for preview

bothdf.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 

# Convert to DataFrame

# Preview DataFrame
```

```python
tumormean = bothdf['Tumor Volume (mm3)'].mean()
groupdf = bothdf.groupby(['Drug','Timepoint'])
```

```python
tumorgroup = groupdf.mean()['Tumor Volume (mm3)']
tumorchart = pd.DataFrame(tumorgroup)

tumorchart.head(15)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint

# Convert to DataFrame

# Preview DataFrame

```

```python
serror = groupdf.sem()['Tumor Volume (mm3)']
```

```python
serrordf = pd.DataFrame(serror)
```

```python
serrordf.head(15)
```

```python
# Minor Data Munging to Re-Format the Data Frames

# Preview that Reformatting worked
```

```python
tmeanformat = tumorchart.reset_index()
```

```python
tmeanpivot = tmeanformat.pivot(index='Timepoint',columns='Drug')['Tumor Volume (mm3)']
```

```python
tmeanpivot.head()
```

```python
serrorformat =serrordf.reset_index()
```

```python
serrorpivot = serrorformat.pivot(index='Timepoint',columns='Drug')['Tumor Volume (mm3)']
```

```python
serrorpivot.head()
```

```python
# Generate the Plot (with Error Bars)

# Save the Figure

```

```python
x_axis = np.arange(0,50,10)
x_limit = 50
```

```python

plt.figure(figsize=(6,4))

plt.errorbar(tmeanpivot.index, tmeanpivot["Capomulin"], yerr=serrorpivot["Capomulin"],
             color="r", marker="o", markersize=5, linewidth=1)
plt.errorbar(tmeanpivot.index, tmeanpivot["Infubinol"], yerr=serrorpivot["Infubinol"],
             color="b", marker="o", markersize=5, linewidth=1)
plt.errorbar(tmeanpivot.index, tmeanpivot["Ketapril"], yerr=serrorpivot["Ketapril"],
             color="g", marker="o", markersize=5, linewidth=1)
plt.errorbar(tmeanpivot.index, tmeanpivot["Naftisol"], yerr=serrorpivot["Naftisol"],
             color="k", marker="o", markersize=5, linewidth=1)

plt.title("Tumor Response to Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Tumor Volume (mm3)")
plt.legend(labels=['Capomulin','Infubinol','Ketapril','Naftisol'],loc='best', fancybox=True)

plt.grid(linestyle="dashed")



plt.savefig('TumorChart')
plt.show()


```

```python
changetime = (tmeanpivot.iloc[-1] - tmeanpivot.iloc[0])/tmeanpivot.iloc[0]*100
#errortime = (serrorpivot.iloc[-1] - serrorpivot.iloc[0])/serrorpivot.iloc[0]*100
```

```python
# Store all Relevant Percent Changes into a Tuple
storechanges = (changetime["Capomulin"], 
                changetime["Infubinol"], 
                changetime["Ketapril"], 
                changetime["Placebo"])

# Splice the data between passing and failing drugs



#I Followed this documentation but i didnt really understand https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots.html

fig, ax = plt.subplots()
lenth = np.arange(len(storechanges))  
width = .75
bpass = ax.bar(lenth[0], storechanges[0], width, color='green')
bfail = ax.bar(lenth[1:], storechanges[1:], width, color='red')




# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_xticks(lenth)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
ax.set_ylim([-20,65])




# Use functions to label the percentages of changes
def perchange():
    for p in ax.patches:
        width, height = p.get_width(), p.get_height()
        x, y = p.get_xy() 
        ax.annotate('{:.0%}'.format(height/100), (x, y + height))

# Call functions to implement the function calls

perchange()
# Save the Figure
fig.savefig("barcharttumor.png")

# Show the Figure
fig.show()
```
